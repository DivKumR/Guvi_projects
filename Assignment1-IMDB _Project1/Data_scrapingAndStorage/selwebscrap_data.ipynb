```python
# Install selenium
%pip install selenium

# Import external modules and libraries for selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Import external modules and libraries for python
import pandas as pd
%pip install pandas
import glob
```

```python
""" !*** -- NOTE: UnComment for Each Genre and run the code. (Select multiple lines and press ctrl + / to comment or umcomment multiple lines)
#Example: For Collecting datas for genre action movies 
1.Uncomment webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action');
2. Uncomment creating data frame step and save to csv step # # Save extracted lists to data frame and save data as csv '<genres>.csv'
-- ***! """

# Open imdb url for different genres in chrome.

driver = webdriver.Chrome()
driver.maximize_window()

# # NOTE: UnComment and comment webpage_url for Each Genre as needed and run the code.
# webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action');
# webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=comedy');
# webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=animation');
# webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=crime');
webpage_url=('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=romance');
driver.get(webpage_url)
time.sleep(5)

# Import libraries
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Click '50 more' button 4 times, this allows to view 200 datas in the page
for _ in range(3):
    try:
        # Locate the "50 more" button
        see_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, "ipc-see-more__text"))
        )
        
        # Scroll into view before clicking
        ActionChains(driver).move_to_element(see_more_button).perform()
        
        # JavaScript click in case Selenium click gets intercepted
        driver.execute_script("arguments[0].click();", see_more_button)
        time.sleep(5)  # Wait for page to load

    except Exception as e:
        print(f"Error clicking '50 more': {e}")
        break  # Stop loop if button is missing or fails

# Create an empty list to store titles
titles_list = []
rating_list=[]
vote_count=[]
movie_duration=[]

# Find all movie items using Xpath
movie_items = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')

# Loop through each movie item for finding movie title, vote,rating and duration using XPATH
for movie_item in movie_items:
    try:
        # Locate the title element within each movie item
        title_element = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[1]/a/h3')
        # trimmed_title = title_element.text.split(". ", 1)[-1]  # Removes everything before ". " including the dot and space

        # Locate the rating element within each movie item
        rating_element = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[1]')

        # Locate the vote element within each movie item
        vote_element = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[2]')

        # Locate the duration element within each movie item
        duration_element = movie_item.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[2]/span[2]')

        # Append the extracted text to the respective list
        titles_list.append(title_element.text.split(". ", 1)[-1]) # Removes everything before ". " including the dot and space
        rating_list.append(rating_element .text)
        vote_count.append(vote_element.text.strip(" ()")) # Removes parentheses
        movie_duration.append(duration_element.text)

        # Append None if the element is not found
    except:
        titles_list.append(None)
        rating_list.append(None)
        vote_count.append(None)
        movie_duration.append(None)

# Print the stored list
print(titles_list)
print(rating_list)
print(vote_count)
print(movie_duration)


""" # Genre-wise Storage: Save extracted data as individual CSV files for each genre. """
import os
os.chdir("C:/Users/v-dhramaraj/Desktop/Python/Projects/IMDB _Project1/IMDB_Project1_CollectedCsvData") # Define Path to store the csv file
# print(os.getcwd())

# NOTE: Uncomment below code for creating data frame and save to csv step # # Save extracted lists to data frame and save data as csv '<genres>.csv file'

""" ===========================================================
# # Save extracted lists to data frame and save data as csv 'action_movies.csv' """
# # Create Data Frame 

# Action_df = pd.DataFrame({
#     'Title': titles_list,
#     'Rating': rating_list,
#     'Votes': vote_count,
#     'Duration': movie_duration
# })
# # Add Genre column with 'Action' as default value
# Action_df["Genre"] = "Action"
# Action_df.to_csv(r"action_movies.csv", index=False) #index=False argument prevents pandas from writing the index column into the CSV file

""" ===========================================================
# # Save extracted lists to data frame and save data as csv 'comedy_movies.csv' """
# # Create Data Frame 

# Comedy_df = pd.DataFrame({
#     'Title': titles_list,
#     'Rating': rating_list,
#     'Votes': vote_count,
#     'Duration': movie_duration
# })
# # Add Genre column with 'Comedy' as default value
# Comedy_df["Genre"] = "Comedy"

# Comedy_df.to_csv(r"comedy_movies.csv", index=False)


""" ===========================================================
# # Save extracted lists to data frame and save data as csv 'Animation_movies.csv' """
# # Create Data Frame 
# Animation_df = pd.DataFrame({
#     'Title': titles_list,
#     'Rating': rating_list,
#     'Votes': vote_count,
#     'Duration': movie_duration
# })
# # Add Genre column with 'Animation' as default value
# Animation_df["Genre"] = "Animation"

# Animation_df.to_csv(r"Animation_movies.csv", index=False)

""" ===========================================================
# # Save extracted lists to data frame and save data as csv 'Crime_movies.csv' """

# # Create Data Frame 
# Crime_df = pd.DataFrame({
#     'Title': titles_list,
#     'Rating': rating_list,
#     'Votes': vote_count,
#     'Duration': movie_duration
# })
# # Add Genre column with 'Crime' as default value
# Crime_df["Genre"] = "Crime"

# Crime_df.to_csv(r"Crime_movies.csv", index=False)

""" ===========================================================
# # Save extracted lists to data frame and save data as csv 'Romance_movies.csv' """

# # Create Data Frame 
# Romance_df = pd.DataFrame({
#     'Title': titles_list,
#     'Rating': rating_list,
#     'Votes': vote_count,
#     'Duration': movie_duration
# })
# # Add Genre column with 'Romance' as default value
# Romance_df["Genre"] = "Romance"

# Romance_df.to_csv(r"Romance_movies.csv", index=False)
```
